```python
import numpy as np
from math import factorial

num_combos = lambda n,k: int(factorial(n)/(factorial(k)*factorial(n-k)))

slow_rows = raw_df[raw_df['TOT_slow']==1].drop(output_cols, axis=1).sample(frac=1).values
n_inputs = slow_rows.shape[1]
n_dists = num_combos(slow_rows.shape[0], 2)

print(f'rows:{slow_rows.shape[0]}, cols:{n_inputs}, combos:{n_dists}')

distances = {x:np.zeros(n_dists) for x in range(n_inputs)}
r = 0
for i in range(slow_rows.shape[0]):
    for j in range(i):
        d_ij = slow_rows[i] - slow_rows[j]
        for x in range(n_inputs):
            distances[x][r] = abs(d_ij[x])
        r += 1
print('distances calculated... computing metrics')
avg_distances = {x:sum(dists)/len(dists) for x,dists in distances.items()}
min_distances = {x:min(dists) for x,dists in distances.items()}
max_distances = {x:max(dists) for x,dists in distances.items()}
print('Avg Distances:', avg_distances)
print('Min Distances:', min_distances)
print('Max Distances:', max_distances)

# from sklearn.metrics.pairwise import euclidean_distances
# dists = euclidean_distances(slow_rows, slow_rows)
```
